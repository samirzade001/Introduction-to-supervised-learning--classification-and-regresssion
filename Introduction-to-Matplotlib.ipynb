## Matplotlib Example Workflow

```python
# using '%' for importing all type of graphs and plot of matplotlib into the notebook
%matplotlib inline
import matplotlib.pyplot as plt  #importing the matplotlib and get ready for plotting.

#creating x and y axis(preparing the data for plotting.)
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#creating the plot for ploting the x and y axis data.
fig, ax = plt.subplots(figsize=(5, 5))

#adding title to plot, x axis and y axis.
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

#plotting the graph on notebook.
ax.plot(x, y);  #using';' for hidding the array information.
```

## Two Options for Subplots

### Option 1 for subplots -

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))


## ploting to each diffrent axis with diffrent types.
ax1.plot(x, x/2);                                               # LinePlot
ax2.scatter(np.random.random(10), np.random.random(10));        # Scatter Plot
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());  # Bar Plot
ax4.hist(np.random.randn(1000));                                # Histogram
```

### Option 2 for subplots -

```python
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which One should we used? (pyplot or Matplotlib OO Method?)
* When to plotting something quickly, used pyplot method.
* When to plotting something in advanced, used Matplotlib OO Method.

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
## PLotting with the help of "Pyplot Method."
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
             figsize = (10, 6));
```

```python
## Plotting with help of "Matplotlib OO Method" mixed with the pyplot method.
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target",
             ax = ax);
```

```python
## Plotting with help of "Matplotlib OO Method".
fig, ax = plt.subplots(figsize = (10, 6))

## Plot the graph.
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

## Customize the plot.
ax.set(title = "Heart Disease and Cholestrol Level",
       xlabel = "Age",
       ylabel = "Cholestrol");

## Add a legend.
ax.legend(*scatter.legend_elements(), title = "target");

## add a horizontal mean line.
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
## Subplots of 'chol', 'age', and 'thalach'
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 10))

## Add data to 'ax0'
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

## customize the 'ax0' plots.
ax0.set(title = "Heart Disease and Cholestrol Levels",
        xlabel = "Age",
        ylabel = "Cholestrol")

## Add a legend to 'ax0'
ax0.legend(*scatter.legend_elements(), title = "Target")

## Add a meanline to the plot.
ax0.axhline(y = over_50["chol"].mean(),
            linestyle = "--")

# --------------------------------------------------------------------------------------------------#

## add data to 'ax1'.
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

## customize the 'ax1' plot.
ax1.set(title = "Heart Disease and Max Heart Rate.",
        xlabel = "age",
        ylabel = "Heart Rate")

## add a legend to plot
ax1.legend(*scatter.legend_elements(), title = "Target")

## add a meanline to the plot.
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = "--");

## add a title to the 'entire plot'.
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
## setting the style to the plot.
plt.style.use("seaborn-whitegrid")

## OO method from the scratch.
fig, ax = plt.subplots(figsize = (10, 4))

## Plot the graph.
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter")   ## 'cmap' used to change the theme of plot with unique colors.

## Customize the plot.
ax.set(title = "Heart Disease and the Cholestrol Level Analysis",
       xlabel = "Age",
       ylabel = "Cholestrol")

## Adding the legends to the plot.
ax.legend(*scatter.legend_elements(), title = "Target")

## Add a horizontal line to the plot.
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
## customizing the y and x limitations.

## Subplots of 'chol', 'age', and 'thalach'
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 10))

## Add data to 'ax0'
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"],
                      cmap = "winter")

## customize the 'ax0' plots.
ax0.set(title = "Heart Disease and Cholestrol Levels",
        xlabel = "Age",
        ylabel = "Cholestrol")

## change the x axis limits.
ax0.set_xlim([50, 80])

## Add a legend to 'ax0'
ax0.legend(*scatter.legend_elements(), title = "Target")

## Add a meanline to the plot.
ax0.axhline(y = over_50["chol"].mean(),
            linestyle = "--")

# --------------------------------------------------------------------------------------------------#

## add data to 'ax1'.
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = "winter")

## change the x axis limits.
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

## customize the 'ax1' plot.
ax1.set(title = "Heart Disease and Max Heart Rate.",
        xlabel = "age",
        ylabel = "Heart Rate")

## add a legend to plot
ax1.legend(*scatter.legend_elements(), title = "Target")

## add a meanline to the plot.
ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = "--");

## add a title to the 'entire plot'.
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
